```python
import pandas as pd
import numpy as np
```

```python
NBA_dict = {'player' : ['Stephen Curry', 'Bradley Beal', 'Damian Lillard', 'Joel Embiid', 'Giannis Antetokounmpo'],
              'team' : ['GSW', 'WAS', 'POR', 'PHI', 'MIL'],
              'points/game' : [32, 31.3, 28.8, 28.5, 28.1]}
```

```python
NBA = pd.DataFrame(NBA_dict)
```

```python
NBA['min'] = [34.2, 35.8, 35.8, 31.1, 33]
```

```python
NBA.loc[4,'min'] = np.NaN
```

```python
NBA.info()  #missing value will be null in info
```

```python
NBA.corr()  #correlation between points/game and min
```

```python
NBA.cov()  #covariance between points/game and min
```

```python
NBA['points/game'].quantile(0.60)
```
